```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
```

```python
class Batsman(TypedDict):
    runs:int
    balls:int
    fours:int
    sixes:int
    strike:float
    bpb:float
    boundary_percent:float
    summary:str

```

```python
graph=StateGraph(Batsman)
```

```python
def calculate_boundry_percent(state:Batsman) -> Batsman:
    boundary_percent = (((state['fours'] * 4) + (state['sixes'] * 6))/state['runs'])*100

    return {'boundary_percent': boundary_percent}
    
```

```python
def calculate_bpb(state:Batsman) -> Batsman:
    bpb = state['balls']/(state['fours'] + state['sixes'])

    return {'bpb': bpb}

```

```python
def calculate_strike_rate(state:Batsman) -> Batsman:
    strike = (state['runs']/state['balls'])*100
    return {'strike': strike}
    
```

```python
def summary(state:Batsman)->Batsman:
    summary = f"""
Strike Rate - {state['strike']} \n
Balls per boundary - {state['bpb']} \n
Boundary percent - {state['boundary_percent']}
"""
    
    return {'summary': summary}
```

```python
graph.add_node('calculate_boundry_percent',calculate_boundry_percent)
graph.add_node('calculate_bpb',calculate_bpb)
graph.add_node('calculate_strike_rate',calculate_strike_rate)
graph.add_node('calculate_summary',summary)
```

```python
graph.add_edge(START,'calculate_boundry_percent')
graph.add_edge(START,'calculate_bpb')
graph.add_edge(START,'calculate_strike_rate')
graph.add_edge('calculate_boundry_percent','calculate_summary')
graph.add_edge('calculate_bpb','calculate_summary')
graph.add_edge('calculate_strike_rate','calculate_summary')
graph.add_edge('calculate_summary')
```
